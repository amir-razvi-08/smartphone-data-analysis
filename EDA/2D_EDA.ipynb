```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

### categorical Column
`brand`, `os`, `processor_name`, `has_5G`, `has_NFC`, `has_IR_Blaster`, `fast_charging`, `dual_display`, `card_support`, `model`

### Numerical Column
`price`, `rating`, `processor_cores`, `ram`, `internal_memory`, `battery_mah`, `screen_size`, `ppi`, `refresh_rate`, `num_cameras`, `rear_camera`, `front_camera`

```python
df=pd.read_csv('../dataProcessing/smartphones.csv')
```

### 1. Average Price by Brand

```python
x = df.groupby('brand').count()['model']
temp_df = df[df['brand'].isin(x[x > 10].index)]

plt.figure(figsize=(15, 8))
sns.barplot(
    data=temp_df,
    x='brand',
    y='price',
    hue='brand',
    errorbar=None,
    legend=False,
    palette='crest'
)
plt.xticks(rotation=30)
plt.title('Average Smartphone Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price (INR)')
plt.show()
```

**Insight:**  
The analysis shows a clear distinction between premium and budget smartphone brands:  

- **Apple** stands out with the highest average price, often exceeding **1,00,000**, reflecting its premium market position.  
- **Google** devices follow, averaging around **60,000**, targeting the upper mid-range segment.  
- **Samsung** and **OnePlus** maintain an average price of roughly **40,000**, balancing flagship and affordable models.

Overall, the data highlights a **tiered pricing structure**, where Apple dominates the luxury range,
while brands like Samsung and OnePlus appeal to users seeking flagship-level features at relatively lower prices.

### 2. Operating System vs Price

On the other hand, **Android** is used in all price levels - from budget to flagship.  
Thats why the **average Android phone price is much lower** than iOS devices.

### 4. Processor Type vs Price

```python
plt.figure(figsize=(15, 8))
sns.barplot(
    data=df,
    x='processor_name',
    y='price',
    hue='processor_name',
    errorbar=None,
    legend=False,
    palette='crest'
)
plt.xticks(rotation=30)
plt.title('Average Smartphone Price by Processor Type')
plt.xlabel('Processor Name')
plt.ylabel('Average Price (INR)')
plt.show()
```

**Insight:**

**Tensor** and **Snapdragon** chips come next, powering most upper mid-range and flagship devices.

### 5. Processor Cores vs Price

### 22. HeatMap

```python
num_cols = ['price', 'ram', 'internal_memory', 'battery_mah', 
            'processor_speed', 'screen_size', 'ppi', 'refresh_rate', 'rear_camera', 'front_camera']
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Numeric Feature Correlation')
plt.show()
```

**Insight:**  
From the correlation heatmap, we can see that **processor speed**, **internal memory**, and **RAM**  
have the **strongest positive correlation** with price. This means phones with faster processors  
and higher memory configurations are generally more expensive.  

Camera resolution, screen size, and refresh rate show **only moderate or weak correlation** with price,  
and battery capacity has almost **no direct impact**.  

Overall, the **performance-related features** (processor and memory) contribute more to pricing  
than display or battery specifications.


---

## Conclusion (2D EDA)

From the 2D analysis, it becomes clear which specifications have the strongest impact on smartphone pricing. Features related to **performance**, such as **processor speed**, **RAM**, and **internal storage**, show a noticeable positive relationship with price. Phones with faster chipsets and higher memory configurations are consistently positioned in the mid-range and flagship categories.

Display and camera features also influence price, but the relationship is not as direct. For example, higher **refresh rates** and **higher rear camera megapixel counts** are more common in premium devices, yet there are still many affordable phones offering similar values in these areas. This suggests that these features are often used to enhance user experience rather than define the price tier. Meanwhile, **battery capacity** shows almost no direct correlation with price, indicating that it has become a standard expectation across all budgets.

Connectivity and special features also reflect market positioning. Phones with **5G** and **NFC** tend to fall in mid-range and premium segments, while **IR blasters** and **expandable storage** are more commonly found in budget and mid-range devices. This shows a shift where higher-end manufacturers focus on sleek design and fixed storage, whereas budget brands emphasize flexibility and utility.

Overall, the 2D EDA shows that **pricing is primarily driven by performance hardware**, while cameras, displays, and additional features help differentiate user experience rather than strictly determine cost.


---

## Final Conclusion

The analysis shows that many smartphone specifications have become standard across the current market. Features such as 8GB RAM, 128256GB storage, 45005000 mAh battery capacity, and 4864MP rear cameras appear frequently across multiple price segments, especially in mid-range devices. Because these features are widely available, they do not strongly determine the price of a smartphone on their own.

However, the study highlights that **performance-oriented components** have the strongest influence on pricing. Devices with faster processors, more RAM, and larger internal storage are consistently priced higher, reflecting the importance of smooth performance and long-term usability in consumer buying decisions. Display and camera specifications do contribute to price differences, but their impact is more moderate and varies across models. Meanwhile, battery capacity shows minimal correlation with price, suggesting it has become a standard expectation.

### Key Factors That Influence Smartphone Price:
- **Processor performance** (speed and chipset class)
- **RAM capacity** (ability to handle multitasking)
- **Internal storage size** (higher storage aligns with higher pricing)
- **Display features** such as refresh rate and sharpness (secondary influence)
- **Rear camera resolution** (stronger effect in premium models)
- **Connectivity features** (5G and NFC appear more in higher-end devices)

Overall, **performance specifications are the strongest predictors of price**, while display, camera, and additional features enhance user experience but do not solely define the pricing tier. These findings are valuable for developing a price prediction model, ensuring that the model places greater weight on performance-related attributes.
