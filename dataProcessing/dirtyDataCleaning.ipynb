```python
df=smartphones.copy()
```

```python
df['os'].fillna('No',inplace=True)
```

```python
df['price']=df['price'].str.replace('₹','').str.replace(',','').astype('int')
```

```python
df=df[df['price'] >4000]
```

```python
df.rename(columns={'Unnamed: 0':'index'},inplace=True)
```

```python
df.set_index('index',inplace=True)
```

```python
df[~df['battery'].str.contains('Battery')]
```

```python
df[~df['processor'].str.contains('Core')]
```

```python
df[~df['ram'].str.contains('RAM')]
```

```python
df[~df['display'].str.contains('Display')]
```

```python
df[~df['camera'].str.contains('MP')]
```

### solving 2,143,723 display info instead of battry info

```python
df1=df[df.index.isin([2,143,723])]
```

```python
y=df1.iloc[:,6:10].shift(1,axis=1)
```

```python
df.loc[df1.index,df1.columns[6:]]=y
```

```python
df.loc[2,'battery']="3692 mAh Battery with 40W Fast Charging"
df.loc[143,'battery']="3692 mAh Battery with 40W Fast Charging"
df.loc[723,'battery']="2815 mAh Battery with 20W Fast Charging"
```

### shifting column data dual_display phone

```python
df2=df[df.index.isin([196,214,303,351,402,409,410,428,479,502,519,527,550,556,560,604,625,631,636,675,692,794,820,840,859,882,883,931,946,1005])]
```

### shifting card_column data with os column

```python
card_set = set(df[~df['card'].str.contains('Memory')].index)
```

```python
df3=df[df.index.isin(card_set)]
```

```python
card_x=df3.iloc[:,9:].shift(1,axis=1)
```

```python
df.loc[df3.index,df3.columns[9:]]=card_x
```

```python
df['card'].fillna("Memory Card Not Supported",inplace=True)
```

### solving OS column

```python
df['os'].fillna('android',inplace=True)
```

```python
df['os']=df['os'].str.strip().str.split().str.get(0).str.lower()
```
