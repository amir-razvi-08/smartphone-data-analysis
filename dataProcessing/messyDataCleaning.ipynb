```python
import numpy as np
import pandas as pd
```

```python
phones=pd.read_excel('phones.xlsx')
```

### messy DATA
- `sim` contains has 5g and has nfc and has ir_blaster
- `processor` contains processor name, cores info and processor speed
- `ram` contains ram and rom info
- `display` contains screen_size, has dual_display, resolution and refreshrate
- `camera` contains rear_camra and front camera info

```python
phones.head()
```

### filter brand name

```python
phone_brand=phones['model'].str.split(' ').str.get(0).str.lower()
```

```python
phones.insert(1,'brand',phone_brand)
```

```python
phones['model']=phones['model'].str.strip().str.split('(').str.get(0).str.strip()
```

### filter sim column

```python
has_5g=phones['sim'].str.contains('5G')
has_nfc=phones['sim'].str.contains('NFC')
has_ir_blaster=phones['sim'].str.contains('IR Blaster')
```

```python
phones.insert(6,'has_5G',has_5g)
phones.insert(7,'has_NFC',has_nfc)
phones.insert(8,'has_IR_Blaster',has_ir_blaster)
```

```python
phones.head()
```

### filter processor brand

```python
processor_brand=phones['processor'].str.strip().str.split(' ').str.get(0).str.lower()
```

```python
phones.insert(9,'processor_name',processor_brand)
```

```python
phones.loc[phones['processor_name']=='apple','processor_name']='bionic'
phones.loc[phones['processor_name']=='a13','processor_name']='bionic'
```

```python
phones.loc[phones['processor_name']=='google','processor_name']='tensor'
```

```python
phones.loc[phones['processor_name'] == 'tiger', 'processor_name'] = 'unisoc'
phones.loc[phones['processor_name'] == 'octa', 'processor_name'] = 'unisoc'
```

```python
phones.loc[phones['processor_name'] == 'helio', 'processor_name'] = 'dimensity'
phones.loc[phones['processor_name'] == 'dimensit', 'processor_name'] = 'dimensity'
```

 ### solve memory column

```python
phones.rename(columns={"ram":"memory"},inplace=True)
```

```python
ram=phones['memory'].str.strip().str.split(' ').str.get(0).astype('float')
rom=phones['memory'].str.strip().str.split(' ').str.get(3).astype('float')
```

```python
phones.insert(12,'ram',ram)
phones.insert(13,'internal_memory',rom)
```

```python
phones.loc[phones['ram']==128,['ram','internal_memory']]=[8,128]
phones.loc[phones['ram']==256,['ram','internal_memory']]=[8,256]
phones.loc[phones['ram']==512,['ram','internal_memory']]=[8,512]
```

```python
phones.loc[phones['internal_memory']==1,'internal_memory']=1024
phones.loc[phones['internal_memory']==2,'internal_memory']=2048
```

```python
phones.head()
```

### solve battery info

### solving display column

```python
screen_size=phones['display'].str.strip().str.split(',').str.get(0).str.strip().str.split(' ').str.get(0).astype('float')
```

```python
phones.insert(18,'screen_size',screen_size)
```

```python
resolution=phones['display'].str.strip().str.split(',').str.get(1).str.strip().str.split('px').str.get(0)
```

```python
refresh_rate=phones['display'].str.strip().str.split(',').str.get(2).str.strip().str.split(' ').str.get(0).astype('float')
```

```python
phones.head()
```

### solve camera column problem

```python
def camerainfo(text):
    if 'Quad' in text:
        return 5 
    elif 'Triple' in text:
        return 3
    elif 'Dual' in text:
        return 2
    else:
        return 1
```

```python
camera_info=phones['camera'].str.strip().str.split('&').str.get(0).apply(camerainfo)
```

```python
phones.insert(23,'num_rear_cameras',camera_info)
```

```python
rear_camera=phones['camera'].str.strip().str.split(' ').str.get(0).astype('float')
```

```python
phones.insert(24,'rear_camera',rear_camera)
```

```python
front_camera=phones['camera'].str.strip().str.split('&').str.get(1).str.strip().str.split(' ').str.get(0)
```

```python
phones.insert(25,'front_camera',front_camera)
```

```python
phones.loc[phones['front_camera']=='Main','front_camera']='8'
```

```python
phones.head()
```

### filter processor core processor speed

```python
def get_cores(text):
    if 'Octa Core' in text:
        return 8
    elif 'Hexa Core' in text:
        return 6
    elif 'Deca Core' in text:
        return 10
    elif 'Nine Cores' in text or 'Nine-Cores' in text or 'Nine Core' in text :
        return 9
    elif 'Quad Core' in text:
        return 4
```

```python
cores=phones['processor'].str.strip().apply(get_cores)
```

```python
phones.insert(10,'processor_cores',cores)
```

```python
phones['processor_cores'].fillna(8,inplace=True)
```

```python
hz=phones['processor'].str.strip().str.split(',').str.get(2).str.strip().str.split(' ').str.get(0).astype('float')
```

```python
phones.insert(11,'processor_speed',hz)
```

```python
phones.head()
```

### inset card_support column

```python
card=phones['card'].apply(lambda x: False if x == 'Memory Card Not Supported' else True)
```

```python
phones.insert(28,'card_support',card)
```

### feature eng-> insert ppi column

```python
width_pixels=phones['resolution'].str.strip().str.split(' ').str.get(0).astype('float')
```

```python
height_pixels=phones['resolution'].str.strip().str.split(' ').str.get(-1).astype('float')
```

```python
ppi = round(np.sqrt(width_pixels**2 + height_pixels**2) / phones['screen_size']).astype('int')
```

```python
phones.insert(22,'ppi',ppi)
```

```python
phones.drop(columns=['index','sim','battery','display','memory','processor','card','camera','resolution'],inplace=True)
```

```python
phones.to_csv('smartphones.csv',index=False)
```
